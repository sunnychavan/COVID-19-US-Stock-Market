## creating important stock dataframes for 2019 and 2020

```python
#Tech = Apple (AAPL), Amazon (AMZN) and Google (GOOG)     
FN_AAPL_2019 = 'AAPL_2019.csv'
AAPL_2019_df = pd.read_csv(FN_AAPL_2019)

FN_AAPL_2020 = 'AAPL_2020.csv'
AAPL_2020_df = pd.read_csv(FN_AAPL_2020)

FN_AMZN_2019 = 'AMZN_2019.csv'
AMZN_2019_df = pd.read_csv(FN_AMZN_2019)

FN_AMZN_2020 = 'AMZN_2020.csv'
AMZN_2020_df = pd.read_csv(FN_AMZN_2020)

FN_GOOG_2019 = 'GOOG_2019.csv'
GOOG_2019_df = pd.read_csv(FN_GOOG_2019)

FN_GOOG_2020 = 'GOOG_2020.csv'
GOOG_2020_df = pd.read_csv(FN_GOOG_2020)

#Retail = Costco (COST), Walmart (WMT) and Target (TGT) 

FN_COST_2019 = 'COST_2019.csv'
COST_2019_df = pd.read_csv(FN_COST_2019)

FN_COST_2020 = 'COST_2020.csv'
COST_2020_df = pd.read_csv(FN_COST_2020)

FN_WMT_2019 = 'WMT_2019.csv'
WMT_2019_df = pd.read_csv(FN_WMT_2019)

FN_WMT_2020 = 'WMT_2020.csv'
WMT_2020_df = pd.read_csv(FN_WMT_2020)

FN_TGT_2019 = 'TGT_2019.csv'
TGT_2019_df = pd.read_csv(FN_GOOG_2019)

FN_TGT_2020 = 'TGT_2020.csv'
TGT_2020_df = pd.read_csv(FN_TGT_2020)

#Airline = Southwest (SW), delta (Delta), United (UAL)
SW_2019_df = pd.read_csv('SW_2019.csv')
SW_2020_df = pd.read_csv('SW_2020.csv')
Delta_2019_df = pd.read_csv('Delta_2019.csv')
Delta_2020_df = pd.read_csv('Delta_2020.csv')
UAL_2019_df = pd.read_csv('UAL_2019.csv')
UAL_2020_df = pd.read_csv('UAL_2020.csv')

#Pharma= Johnson and Johnson (JNJ), Merck (Merck), Pfizer (Pfizer)
JNJ_2019_df = pd.read_csv('JNJ_2019.csv')
JNJ_2020_df = pd.read_csv('JNJ_2020.csv')
Merck_2019_df = pd.read_csv('Merck_2019.csv')
Merck_2020_df = pd.read_csv('Merck_2020.csv')
Pfizer_2019_df = pd.read_csv('Pfizer_2019.csv')
Pfizer_2020_df = pd.read_csv('Pfizer_2020.csv')

#SPY

SPY_2019_df = pd.read_csv('SPY_2019.csv')
SPY_2020_df = pd.read_csv('SPY_2020.csv')





```

```python
#creating adjusted closing price series for each company for each year 
#tech 
Pct_AAPL_2019 = AAPL_2019_df['Adj Close'].pct_change()
Pct_AMZN_2019 = AMZN_2019_df['Adj Close'].pct_change()
Pct_GOOG_2019 = GOOG_2019_df['Adj Close'].pct_change()
Pct_AAPL_2020 = AAPL_2020_df['Adj Close'].pct_change()
Pct_AMZN_2020 = AMZN_2020_df['Adj Close'].pct_change()
Pct_GOOG_2020 = GOOG_2020_df['Adj Close'].pct_change()

#retail
Pct_COST_2019 = COST_2019_df['Adj Close'].pct_change()
Pct_WMT_2019 = WMT_2019_df['Adj Close'].pct_change()
Pct_TGT_2019 = TGT_2019_df['Adj Close'].pct_change()
Pct_COST_2020 = COST_2020_df['Adj Close'].pct_change()
Pct_WMT_2020 = WMT_2020_df['Adj Close'].pct_change()
Pct_TGT_2020 = TGT_2020_df['Adj Close'].pct_change()

#Airlines
Pct_SW_2019 = SW_2019_df['Adj Close'].pct_change()
Pct_Delta_2019 = Delta_2019_df['Adj Close'].pct_change()
Pct_UAL_2019 = UAL_2019_df['Adj Close'].pct_change()
Pct_SW_2020 = SW_2020_df['Adj Close'].pct_change()
Pct_Delta_2020 = Delta_2020_df['Adj Close'].pct_change()
Pct_UAL_2020 = UAL_2020_df['Adj Close'].pct_change()

#Pharma
Pct_JNJ_2019 = JNJ_2019_df['Adj Close'].pct_change()
Pct_Merck_2019 = Merck_2019_df['Adj Close'].pct_change()
Pct_Pfizer_2019 = Pfizer_2019_df['Adj Close'].pct_change()
Pct_JNJ_2020 = JNJ_2020_df['Adj Close'].pct_change()
Pct_Merck_2020 = Merck_2020_df['Adj Close'].pct_change()
Pct_Pfizer_2020 = Pfizer_2020_df['Adj Close'].pct_change()

#SPY
Pct_SPY_2019 = SPY_2019_df['Adj Close'].pct_change()
Pct_SPY_2020 = SPY_2020_df['Adj Close'].pct_change()
 
```

```python
#dwy = "date without year"Pct_APPL_2019 = AAPL_2019_df['Adj Close'].pct_change()


#creating lists of dates without year so we can compare stock prices for both years 

numpy_dates = AAPL_2019_df['Date'].values  
dwy = []
for x in range(len(numpy_dates)):
    dwy.append(numpy_dates[x][:-5])
    




```

```python
# pd.pct_change (dataframe of percentage change)
#creating a function that graphs percent change between 2019 and 2020 for a given company 

def Pct_change_grapher(dataframe_2019,dataframe_2020,date_list, company_name):
    
    Pct_df_2019 = dataframe_2019['Adj Close'].pct_change()
    Pct_df_2020 = dataframe_2020['Adj Close'].pct_change()
   
    #plot syntax
    fig, axes = plt.subplots(nrows=1, ncols=1)
    
    axes.plot(date_list,Pct_df_2019)
    axes.plot(date_list,Pct_df_2020)
    axes.xaxis.set_major_locator(plt.MaxNLocator(11))
    axes.tick_params(axis = 'x', rotation = 30 )
    plt.legend(['2019',"2020"])
    axes.set_xlabel("date")
    axes.set_ylabel("percent change")
    axes.set_title("Comparing " + company_name + ' percent change in stock for 2019 and 2020 between 1/2 and 4/17')
    plt.axhline(linewidth=1, color='k')

    



```
